```python
import numpy as np
import pandas as pd
import os
from itertools import combinations
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw

def load_trajectories(base_path, robot_ids, num_runs):
    # Load all robot trajectories (list of (x,y) points)
    trajectories = {r: {} for r in robot_ids}
    for r in robot_ids:
        for run in range(1, num_runs+1):
            file_path = os.path.join(base_path, f"Data{run}", f"tb3_{r}.csv")
            if os.path.isfile(file_path):
                data = pd.read_csv(file_path)
                traj = list(zip(data[' x'], data[' y']))
                trajectories[r][run] = traj
    return trajectories

def compute_mean_intra_dtw_normalized(trajectories, robot_ids, num_runs, workspace_extent, resample_trajectory, target_length):
    """
    Compute mean, variance, and SD of intra-robot DTW distances,
    NORMALIZED by the workspace extent.
    """
    # Prevent division by zero if workspace is empty
    if workspace_extent == 0:
        workspace_extent = 1.0

    mean_dtw_per_robot = []
    var_dtw_per_robot = []
    sd_dtw_per_robot = []
    all_dtw_normalized = []

    for r in robot_ids:
        dtw_vals_normalized = []
        available_runs = [run for run in range(1, num_runs + 1) if run in trajectories[r]]
        for a, b in combinations(available_runs, 2):
            traj_a = resample_trajectory(trajectories[r][a], target_length)
            traj_b = resample_trajectory(trajectories[r][b], target_length)
            
            # Calculate raw DTW
            d, _ = fastdtw(traj_a, traj_b, dist=euclidean)
            
            # Normalize the DTW score
            d_normalized = d / workspace_extent
            
            dtw_vals_normalized.append(d_normalized)
            all_dtw_normalized.append(d_normalized)

        if dtw_vals_normalized:
            mean_dtw_per_robot.append(np.mean(dtw_vals_normalized))
            var_dtw_per_robot.append(np.var(dtw_vals_normalized))
            sd_dtw_per_robot.append(np.std(dtw_vals_normalized))
        else:
            mean_dtw_per_robot.append(np.nan)
            var_dtw_per_robot.append(np.nan)
            sd_dtw_per_robot.append(np.nan)

    overall_mean = np.nanmean(mean_dtw_per_robot)
    overall_var = np.nanmean(var_dtw_per_robot)
    overall_sd = np.nanmean(sd_dtw_per_robot)

    print("--- Normalized Intra-Robot DTW Metrics ---")
    print("Per-robot mean normalized DTW:", np.round(mean_dtw_per_robot, 4))
    print("Per-robot Standard Deviation:", np.round(sd_dtw_per_robot, 4))
    print("Overall mean normalized DTW:", round(overall_mean, 4))
    print("Overall variance of normalized DTW:", round(overall_var, 4))
    print("Overall SD of normalized DTW:", round(overall_sd, 4))

    return mean_dtw_per_robot, var_dtw_per_robot, sd_dtw_per_robot, all_dtw_normalized


def compute_workspace_extent(trajectories):
    # Get all x and y from all trajectories
    all_points = []
    for robot_runs in trajectories.values():
        for traj in robot_runs.values():
            all_points.extend(traj)
    all_points = np.array(all_points)
    # Extent is max difference in x and y
    min_xy = np.min(all_points, axis=0)
    max_xy = np.max(all_points, axis=0)
    workspace_extent = euclidean(min_xy, max_xy)
    return workspace_extent

def compute_start_end_distances(trajectories, robot_ids, num_runs):
    # For all available (robot, run): get start and end positions
    start_points, end_points = [], []
    for r in robot_ids:
        for run in range(1, num_runs+1):
            traj = trajectories[r].get(run)
            if traj is not None and len(traj) > 0:
                start_points.append(traj[0])
                end_points.append(traj[-1])
    # Compute all pairwise euclidean distances between starts and ends
    all_points = start_points + end_points
    dists = [euclidean(all_points[i], all_points[j])
             for i, j in combinations(range(len(all_points)), 2)]
    mean_dist = np.mean(dists)
    max_dist = np.max(dists)
    return mean_dist, max_dist

def compute_mean_trajectory_length(trajectories, robot_ids, num_runs):
    lengths = []
    for r in robot_ids:
        for run in range(1, num_runs+1):
            traj = trajectories[r].get(run)
            if traj is not None and len(traj) > 1:
                length = sum(euclidean(traj[i], traj[i+1]) for i in range(len(traj)-1))
                lengths.append(length)
    return np.mean(lengths) if lengths else 0

# Example usage assumes you have implemented/resampled trajectories as before:
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
base_path = "/home/sweksha/ros_ws/src/swarm_aggregation/Data/boids"
robot_ids = range(1, 7)
num_runs = 29
target_length = 110

trajectories = load_trajectories(base_path, robot_ids, num_runs)

# 1. First, compute the overall workspace extent
workspace_extent = compute_workspace_extent(trajectories)
print(f"Workspace spatial extent: {workspace_extent:.3f}\n")

# 2. Then, pass it to the normalized DTW function
mean_dtw, var_dtw, sd_dtw, all_dtw = compute_mean_intra_dtw_normalized(
    trajectories, robot_ids, num_runs, workspace_extent, resample_trajectory, target_length
)

# ... The rest of your analysis ...
mean_se_dist, max_se_dist = compute_start_end_distances(trajectories, robot_ids, num_runs)
mean_traj_len = compute_mean_trajectory_length(trajectories, robot_ids, num_runs)

print("\n--- Other Spatial Metrics ---")
print("Mean start/end-point distance:", mean_se_dist)
print("Max start/end-point distance:", max_se_dist)
print("Mean trajectory length:", mean_traj_len)
```

```python
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
import pandas as pd
import numpy as np
import os
from scipy.optimize import linear_sum_assignment

def load_trajectories(run_path, num_robots):
    """
    Load trajectories for all robots in a given run folder.
    
    Args:
        run_path (str): directory path for a single run
        num_robots (int): number of robots
    
    Returns:
        dict: {robot_id: trajectory as list of (x,y)}
    """
    trajectories = {}
    for r in range(1, num_robots + 1):
        file_path = os.path.join(run_path, f"tb3_{r}.csv")
        if os.path.isfile(file_path):
            data = pd.read_csv(file_path)
            trajectories[r] = list(zip(data['x'], data['y']))
        else:
            raise FileNotFoundError(f"Trajectory file missing: {file_path}")
    return trajectories

def compute_reid_accuracy(run1_path, run2_path, num_robots):
    """
    Compute trajectory re-identification accuracy between two runs.
    
    Args:
        run1_path (str): path to first run folder
        run2_path (str): path to second run folder
        num_robots (int): number of robots
    
    Returns:
        float: re-identification accuracy (fraction correct)
    """
    # Load trajectories
    trajs_run1 = load_trajectories(run1_path, num_robots)
    trajs_run2 = load_trajectories(run2_path, num_robots)

    # Compute DTW distance matrix
    dist_matrix = np.zeros((num_robots, num_robots))
    for i in range(1, num_robots + 1):
        traj_i = trajs_run1[i]
        for j in range(1, num_robots + 1):
            traj_j = trajs_run2[j]
            d, _ = fastdtw(traj_i, traj_j, dist=euclidean)
            dist_matrix[i - 1, j - 1] = d

    # Solve assignment problem to minimize total DTW cost
    row_ind, col_ind = linear_sum_assignment(dist_matrix)
    
    # Correct matches count where matched robot indices align
    correct_matches = np.sum(col_ind == np.arange(num_robots))
    
    accuracy = correct_matches / num_robots
    return accuracy

# Example usage:
base_path = "/home/sweksha/Data"
from itertools import combinations

base_run_names = [f"Lidar12_3_{i}" for i in range(1,6)]
run_pairs = list(combinations(base_run_names, 2))

num_robots = 11
accuracies = []
for run1, run2 in run_pairs:
    acc = compute_reid_accuracy(os.path.join(base_path, run1),
                                os.path.join(base_path, run2), num_robots)
    accuracies.append(acc)
mean_acc = np.mean(accuracies)
print(f"Mean trajectory re-ID accuracy: {mean_acc:.3f}, {accuracies}")
```

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import entropy

def compute_kl_between_runs(robot_id, base_path, num_runs, bins=50):
    """
    Compute mean symmetric KL divergence between trajectory distributions for a robot.
    Automatically determines histogram bounds from all runs.
    """
    trajectories = []
    all_points = []
    
    # 1 Load data
    for run in range(1, num_runs + 1):
        file_path = os.path.join(base_path, f"Lidar20_3_{run}", f"tb3_{run}.csv")
        if not os.path.exists(file_path):
            print(f"Warning: {file_path} not found.")
            continue
        data = pd.read_csv(file_path)
        if 'x' not in data.columns or 'y' not in data.columns:
            print(f"Warning: Missing x/y in {file_path}")
            continue
        x, y = data['x'].dropna().values, data['y'].dropna().values
        if len(x) == 0:  # no valid points
            continue
        all_points.append(np.column_stack((x, y)))

    if len(all_points) < 2:
        print(f"Not enough valid runs for robot {robot_id}")
        return np.nan

    # 2 Determine global bounds
    all_xy = np.vstack(all_points)
    x_min, x_max = all_xy[:, 0].min(), all_xy[:, 0].max()
    y_min, y_max = all_xy[:, 1].min(), all_xy[:, 1].max()
    bounds = [[x_min, x_max], [y_min, y_max]]

    # 3 Build normalized histograms
    for pts in all_points:
        hist, _, _ = np.histogram2d(pts[:, 0], pts[:, 1], bins=bins, range=bounds)
        hist = hist.flatten()
        if np.sum(hist) == 0:
            continue
        hist = hist / np.sum(hist)
        trajectories.append(hist)
    
    if len(trajectories) < 2:
        print(f"Insufficient valid histograms for robot {robot_id}")
        return np.nan

    # 4 Compute pairwise symmetric KL
    kl_values = []
    for i in range(len(trajectories)):
        for j in range(i + 1, len(trajectories)):
            p = trajectories[i] + 1e-12
            q = trajectories[j] + 1e-12
            kl_pq = entropy(p, q)
            kl_qp = entropy(q, p)
            kl_sym = 0.5 * (kl_pq + kl_qp)
            kl_values.append(kl_sym)

    return np.mean(kl_values) if kl_values else np.nan


#  Example usage
base_path = "/home/sweksha/Data"
num_runs = 5
robots = range(1, 7)

mean_kl_per_robot = [compute_kl_between_runs(i, base_path, num_runs) for i in robots]
overall_mean_kl = np.nanmean(mean_kl_per_robot)

print("Mean KL divergence per robot:", mean_kl_per_robot)
print("Overall mean KL divergence:", overall_mean_kl)

```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import entropy
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean

# ----------------------------- #
# Helper Functions
# ----------------------------- #
def compute_jsd(p, q):
    """Compute JensenShannon Divergence (base 2)."""
    eps = 1e-10
    p = np.clip(p, eps, 1)
    q = np.clip(q, eps, 1)
    m = 0.5 * (p + q)
    return 0.5 * (entropy(p, m, base=2) + entropy(q, m, base=2))


def compute_jsd_between_runs(robot_id, base_path, num_runs, bins=50):
    """Compute pairwise JSD between all runs for a given robot."""
    trajectories = []
    for run in range(1, num_runs + 1):
        folder = os.path.join(base_path, f"Lidar6_3_{run}")
        file_path = os.path.join(folder, f"tb3_{robot_id}.csv")
        df = pd.read_csv(file_path)
        xy = df[['x', 'y']].to_numpy()
        trajectories.append(xy)

    jsd_matrix = np.zeros((num_runs, num_runs))
    jsd_vals = []

    for i in range(num_runs):
        for j in range(i + 1, num_runs):
            hist_i, _, _ = np.histogram2d(
                trajectories[i][:, 0], trajectories[i][:, 1], bins=bins, density=True
            )
            hist_j, _, _ = np.histogram2d(
                trajectories[j][:, 0], trajectories[j][:, 1], bins=bins, density=True
            )
            p = hist_i.flatten()
            q = hist_j.flatten()
            p /= np.sum(p) if np.sum(p) > 0 else 1
            q /= np.sum(q) if np.sum(q) > 0 else 1
            jsd_val = compute_jsd(p, q)
            jsd_matrix[i, j] = jsd_val
            jsd_matrix[j, i] = jsd_val
            jsd_vals.append(jsd_val)
    return np.array(jsd_vals), np.mean(jsd_vals)


def compute_dtw_between_runs(robot_id, base_path, num_runs):
    """Compute pairwise DTW distance between all runs for a given robot."""
    trajectories = []
    for run in range(1, num_runs + 1):
        folder = os.path.join(base_path, f"Lidar6_3_{run}")
        file_path = os.path.join(folder, f"tb3_{robot_id}.csv")
        df = pd.read_csv(file_path)
        xy = df[['x', 'y']].to_numpy()
        trajectories.append(xy)

    dtw_vals = []
    for i in range(num_runs):
        for j in range(i + 1, num_runs):
            d, _ = fastdtw(trajectories[i], trajectories[j], dist=euclidean)
            dtw_vals.append(d)
    return np.array(dtw_vals), np.mean(dtw_vals)


# ----------------------------- #
# Main Execution
# ----------------------------- #
robots = 6
num_runs = 5
base_path = "/home/sweksha/Data"

jsd_all, dtw_all = [], []
mean_jsd_per_robot, mean_dtw_per_robot = [], []

for robot_id in range(1, robots + 1):
    jsd_vals, mean_jsd = compute_jsd_between_runs(robot_id, base_path, num_runs)
    dtw_vals, mean_dtw = compute_dtw_between_runs(robot_id, base_path, num_runs)

    jsd_all.append(jsd_vals)
    dtw_all.append(dtw_vals)
    mean_jsd_per_robot.append(mean_jsd)
    mean_dtw_per_robot.append(mean_dtw)

# ----------------------------- #
# Visualization
# ----------------------------- #

# Combine for Boxplot Comparison
combined_data = []
for rid in range(robots):
    combined_data.extend([["JSD", rid + 1, v] for v in jsd_all[rid]])
    combined_data.extend([["DTW", rid + 1, v] for v in dtw_all[rid]])

df_combined = pd.DataFrame(combined_data, columns=["Metric", "Robot", "Value"])

# ---- Statistical Boxplot ----
plt.figure(figsize=(10, 6))
sns.boxplot(x="Robot", y="Value", hue="Metric", data=df_combined, palette="Set2")
plt.title("Comparison of JSD and DTW Distributions per Robot")
plt.ylabel("Distance / Divergence Measure")
plt.xlabel("Robot ID")
plt.legend(title="Metric")
plt.tight_layout()
plt.show()

# ---- Summary Bar Chart ----
plt.figure(figsize=(8, 5))
x = np.arange(1, robots + 1)
plt.bar(x - 0.15, mean_dtw_per_robot, width=0.3, label="Mean DTW", color="lightcoral")
plt.bar(x + 0.15, mean_jsd_per_robot, width=0.3, label="Mean JSD", color="skyblue")
plt.xticks(x, [f"Robot {i}" for i in range(1, robots + 1)])
plt.ylabel("Mean Metric Value")
plt.title("Mean DTW vs JSD per Robot")
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------------- #
# Summary Results
# ----------------------------- #
print("Mean DTW per robot:", np.round(mean_dtw_per_robot, 3))
print("Mean JSD per robot:", np.round(mean_jsd_per_robot, 3))
print("Overall mean DTW:", round(np.mean(mean_dtw_per_robot), 3))
print("Overall mean JSD:", round(np.mean(mean_jsd_per_robot), 3))
```

```python
#!/usr/bin/env python3
"""
analyze_clusters_dbscan.py

Reads Data/boids/Data{trial}/ CSV logs, detects clusters with DBSCAN,
computes cluster-wise metrics (compactness, cohesion, dispersion, coverage),
and computes silhouette and Dunn indices per trial. Aggregates across trials
and reports mean +/- 95% CI for each metric.

Assumptions:
- Each trial folder is: <base_path>/Data{trial}/ (trial starts at 1)
- CSV files inside trial folder contain columns with names that include 'x' and 'y'
  (case-insensitive). Example produced by your boids node: "time, x, y".
- Default DBSCAN eps can be tuned (suggest eps  min_spacing * 2 or so).
"""

import os
import glob
import math
import numpy as np
import pandas as pd
from itertools import combinations
from scipy.spatial.distance import pdist, squareform, cdist
from scipy.stats import sem, t
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from scipy.spatial import ConvexHull

# ============ USER CONFIG ==============
base_path = "/home/sweksha/ros_ws/src/swarm_aggregation/Data/Correct_30trails_mine"   # path containing Data1, Data2, ...
num_trials = 29        # number of trials to analyze (1..num_trials)
min_spacing = 0.875        # minimum safe spacing used for reference (meters)
dbscan_eps = 1.70          # DBSCAN eps (meters). Tune for your scenario.
dbscan_min_samples = 2    # minimum samples for DBSCAN cluster
confidence_level = 0.95
output_prefix = "cluster_analysis_results"
# =======================================

# ---------- Utility functions ----------
def find_csv_files(folder):
    """Return list of CSV files in folder that contain x and y columns."""
    files = glob.glob(os.path.join(folder, "*.csv"))
    valid = []
    for f in files:
        try:
            df = pd.read_csv(f, nrows=5)
            cols = [c.lower() for c in df.columns]
            if any(' x' in c for c in cols) and any(' y ' in c for c in cols):
                valid.append(f)
        except Exception:
            continue
    return sorted(valid)

def load_positions_from_folder(folder):
    """Load final positions of all robots in folder.
       Returns Nx2 numpy array of (x,y)."""
    csvs = find_csv_files(folder)
    positions = []
    for f in csvs:
        df = pd.read_csv(f)
        # find x,y column names (case-insensitive)
        col_x = next((c for c in df.columns if ' x' in c.lower()), None)
        col_y = next((c for c in df.columns if ' y ' in c.lower()), None)
        if col_x is None or col_y is None:
            continue
        # use last valid row with finite x,y
        df_valid = df[[col_x, col_y]].dropna()
        if df_valid.shape[0] == 0:
            continue
        x, y = df_valid.iloc[-1].to_numpy(dtype=float)
        if np.isfinite(x) and np.isfinite(y):
            positions.append([x, y])
    if len(positions) == 0:
        return np.empty((0,2))
    return np.array(positions)

def mean_pairwise_distance(points):
    if points.shape[0] < 2:
        return np.nan
    d = pdist(points)
    return float(np.mean(d))

def std_pairwise_distance(points):
    if points.shape[0] < 2:
        return np.nan
    d = pdist(points)
    return float(np.std(d, ddof=0))

def cohesion_to_centroid(points):
    if points.shape[0] == 0:
        return np.nan
    centroid = np.mean(points, axis=0)
    return float(np.mean(np.linalg.norm(points - centroid, axis=1)))

def convex_hull_area(points):
    if points.shape[0] < 3:
        return 0.0
    try:
        hull = ConvexHull(points)
        return float(hull.volume) if hasattr(hull, 'volume') else float(hull.area)
    except Exception:
        return np.nan

def dunn_index(labels, points):
    """
    Dunn index = (min inter-cluster distance) / (max intra-cluster diameter)
    Inter-cluster distance: minimum distance between any two points from different clusters.
    Intra-cluster diameter: maximum pairwise distance within a cluster.
    """
    unique_labels = [lab for lab in np.unique(labels) if lab != -1]
    if len(unique_labels) < 2:
        return np.nan
    clusters = [points[labels == lab] for lab in unique_labels]
    # compute min inter-cluster distance
    min_inter = np.inf
    for a, b in combinations(clusters, 2):
        if a.shape[0] == 0 or b.shape[0] == 0:
            continue
        d = cdist(a, b)
        v = np.min(d)
        if v < min_inter:
            min_inter = v
    # compute max intra-cluster diameter
    max_intra = 0.0
    for c in clusters:
        if c.shape[0] < 2:
            continue
        diam = np.max(pdist(c))
        if diam > max_intra:
            max_intra = diam
    if max_intra == 0.0 or not np.isfinite(min_inter):
        return np.nan
    return float(min_inter / max_intra)

def mean_ci(data, confidence=0.95):
    arr = np.array([x for x in data if (x is not None and not np.isnan(x))], dtype=float)
    if arr.size == 0:
        return (np.nan, np.nan, np.nan)
    m = np.mean(arr)
    sd = np.std(arr, ddof=1) if arr.size > 1 else 0.0
    n = arr.size
    se = sd / math.sqrt(n) if n > 0 else 0.0
    h = se * t.ppf((1 + confidence) / 2, n - 1) if n > 1 else 0.0
    return (m, sd, h)

# ---------- Main analysis ----------
per_trial_summary = []
per_trial_cluster_details = []  # list of per-trial lists of per-cluster dicts

for trial in range(0, num_trials + 1):
    folder = os.path.join(base_path, f"Data{trial}")
    if not os.path.isdir(folder):
        print(f"[WARN] Trial folder missing: {folder}")
        continue

    positions = load_positions_from_folder(folder)
    if positions.shape[0] == 0:
        print(f"[WARN] No positions found in {folder}")
        continue

    # Run DBSCAN
    db = DBSCAN(eps=dbscan_eps, min_samples=dbscan_min_samples).fit(positions)
    labels = db.labels_
    unique_labels = [lab for lab in np.unique(labels) if lab != -1]
    n_clusters = len(unique_labels)

    trial_clusters = []
    # For silhouette we need >=2 clusters and at least 2 samples not noise
    silhouette_val = np.nan
    if n_clusters >= 1 and positions.shape[0] - np.sum(labels == -1) >= 2:
        try:
            silhouette_val = silhouette_score(positions, labels)
        except Exception:
            silhouette_val = np.nan

    # Dunn index
    dunn_val = dunn_index(labels, positions)

    # Compute per-cluster metrics
    for lab in unique_labels:
        pts = positions[labels == lab]
        # exclude clusters with <2 points for pairwise metrics
        compact = mean_pairwise_distance(pts)
        cohesion = cohesion_to_centroid(pts)
        dispersion = std_pairwise_distance(pts)
        coverage = convex_hull_area(pts)
        min_inter_robot = np.nan
        if pts.shape[0] >= 2:
            min_inter_robot = np.min(pdist(pts))
        cluster_dict = {
            'trial': trial,
            'cluster_label': int(lab),
            'n_members': int(pts.shape[0]),
            'compactness': compact,
            'cohesion': cohesion,
            'dispersion': dispersion,
            'coverage': coverage,
            'min_inter_robot': min_inter_robot
        }
        trial_clusters.append(cluster_dict)

    # Aggregate cluster-wise metrics per trial (average across clusters)
    if len(trial_clusters) == 0:
        # no clusters found (all noise) -> treat single cluster over all points?
        # We'll treat whole set as single group for physical metrics (but mark as no clusters)
        compact = mean_pairwise_distance(positions)
        cohesion = cohesion_to_centroid(positions)
        dispersion = std_pairwise_distance(positions)
        coverage = convex_hull_area(positions)
        per_trial_summary.append({
            'trial': trial,
            'n_clusters': 0,
            'n_points': positions.shape[0],
            'avg_compactness': compact,
            'avg_cohesion': cohesion,
            'avg_dispersion': dispersion,
            'avg_coverage': coverage,
            'silhouette': silhouette_val,
            'dunn': dunn_val
        })
    else:
        # compute mean across clusters (unweighted)
        avg_compact = np.nanmean([c['compactness'] for c in trial_clusters])
        avg_cohesion = np.nanmean([c['cohesion'] for c in trial_clusters])
        avg_disp = np.nanmean([c['dispersion'] for c in trial_clusters])
        avg_cov = np.nanmean([c['coverage'] for c in trial_clusters])
        per_trial_summary.append({
            'trial': trial,
            'n_clusters': n_clusters,
            'n_points': positions.shape[0],
            'avg_compactness': avg_compact,
            'avg_cohesion': avg_cohesion,
            'avg_dispersion': avg_disp,
            'avg_coverage': avg_cov,
            'silhouette': silhouette_val,
            'dunn': dunn_val
        })

    per_trial_cluster_details.extend(trial_clusters)
    sil_val = f"{silhouette_val:.3f}" if not np.isnan(silhouette_val) else "nan"
    dunn_val_str = f"{dunn_val:.3f}" if not np.isnan(dunn_val) else "nan"

    print(f"[INFO] Trial {trial}: points={positions.shape[0]}, "
      f"clusters={len(trial_clusters)}, silhouette={sil_val}, dunn={dunn_val_str}")

# Convert to DataFrame and save detailed per-cluster CSV
summary_df = pd.DataFrame(per_trial_summary)
clusters_df = pd.DataFrame(per_trial_cluster_details)
os.makedirs("analysis_outputs", exist_ok=True)
summary_df.to_csv(os.path.join("analysis_outputs", f"{output_prefix}_per_trial_summary.csv"), index=False)
clusters_df.to_csv(os.path.join("analysis_outputs", f"{output_prefix}_per_cluster_details.csv"), index=False)

# Aggregate across trials and compute mean +/- 95% CI
metrics = ['avg_compactness', 'avg_cohesion', 'avg_dispersion', 'avg_coverage', 'silhouette', 'dunn']
agg_results = {}
for m in metrics:
    arr = summary_df[m].values if m in summary_df.columns else np.array([])
    mean, sd, h = mean_ci(arr, confidence=confidence_level)
    agg_results[m] = {'mean': mean, 'sd': sd, 'ci95_halfwidth': h}

print("\n=== AGGREGATED METRICS ACROSS TRIALS ===")
for m, stats in agg_results.items():
    print(f"{m:20s}: mean = {stats['mean']:.4f}, sd = {stats['sd']:.4f}, 95% CI  {stats['ci95_halfwidth']:.4f}")

# Save aggregated results
agg_df = pd.DataFrame([
    {'metric': m, 'mean': v['mean'], 'sd': v['sd'], 'ci95_halfwidth': v['ci95_halfwidth']}
    for m, v in agg_results.items()
])
agg_df.to_csv(os.path.join("analysis_outputs", f"{output_prefix}_aggregated.csv"), index=False)
print(f"\nSaved results to analysis_outputs/ (per-trial, per-cluster, aggregated CSVs).")
```

```python
#!/usr/bin/env python3
import os
import csv
import re

def extract_final_clustering_times(base_dir):
    summary = []
    data_dir = os.path.join(base_dir, "Data", "BeeClust")

    # Loop over all Data{n} directories
    for trial in sorted(os.listdir(data_dir)):
        trial_path = os.path.join(data_dir, trial)
        if not os.path.isdir(trial_path):
            continue

        final_times = []
        for fname in os.listdir(trial_path):
            if re.match(r"tb3_\d+\.csv", fname):
                fpath = os.path.join(trial_path, fname)
                try:
                    with open(fpath, "r") as f:
                        lines = f.readlines()
                        if len(lines) > 1:
                            # Skip header, take last line
                            last_line = lines[-1].strip()
                            time_val = float(last_line.split(",")[0])
                            final_times.append(time_val)
                except Exception as e:
                    print(f"Error reading {fpath}: {e}")

        if final_times:
            final_time = max(final_times)
            summary.append((trial, final_time))
            print(f"{trial}: final clustering time = {final_time:.2f}s")

    # Save summary CSV
    out_path = os.path.join(data_dir, "clustering_summary.csv")
    with open(out_path, "w", newline="") as out_file:
        writer = csv.writer(out_file)
        writer.writerow(["trial", "final_clustering_time"])
        writer.writerows(summary)

    print(f"\n Summary saved to: {out_path}")


if __name__ == "__main__":
    dirname = "/home/sweksha/ros_ws/src/swarm_aggregation"
    extract_final_clustering_times(dirname)
```

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

def parse_swarm_summary(filepath):
    """
    Parses a summary-only cluster_quality.txt file where each algorithm has
    mean/sd values for multiple metrics.
    """
    with open(filepath, 'r') as f:
        content = f.read()

    # Split by algorithm headers like ###PSO###
    blocks = re.split(r'#+\s*([A-Za-z0-9_() -]+)\s*#+', content)
    summary_data = {}

    for i in range(1, len(blocks), 2):
        algo = blocks[i].strip()
        block = blocks[i+1]
        summary_data[algo] = {}

        for line in block.splitlines():
            line = line.strip()
            if not line or ':' not in line:
                continue
            metric = line.split(':')[0].strip()
            mean_match = re.search(r'mean\s*=\s*([-\d.]+)', line)
            sd_match = re.search(r'sd\s*=\s*([-\d.]+)', line)
            if mean_match and sd_match:
                summary_data[algo][metric] = {
                    'mean': float(mean_match.group(1)),
                    'sd': float(sd_match.group(1))
                }

    # Convert to DataFrame for convenience
    flat_data = []
    for algo, metrics in summary_data.items():
        entry = {'Algorithm': algo}
        for metric, vals in metrics.items():
            entry[f'{metric} (mean)'] = vals['mean']
            entry[f'{metric} (sd)'] = vals['sd']
        flat_data.append(entry)

    summary_df = pd.DataFrame(flat_data)

    # --- Synthetic per-trial data for visualization ---
    np.random.seed(42)
    trial_data = []
    for algo, metrics in summary_data.items():
        n_trials = 30  # simulate 30
        for _ in range(n_trials):
            entry = {'Algorithm': algo}
            for metric, vals in metrics.items():
                entry[metric] = max(0, np.random.normal(vals['mean'], vals['sd']))
            trial_data.append(entry)

    trial_df = pd.DataFrame(trial_data)
    return summary_df, trial_df


def calculate_efficiency(df):
    df['Clustering Efficiency'] = df['silhouette'] / df['Clustering time'].replace(0, 1e-6)
    print(df)
    return df


def plot_efficiency_summary(summary_df):
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(10, 6))

    sns.barplot(
    data=summary_df,
    x='Algorithm',
    y='Clustering Efficiency',
    palette='viridis',
    errorbar=None,
    ax=ax
    )
    ax.set_title('Clustering Efficiency (Silhouette / Clustering Time)', fontsize=16, weight='bold')
    ax.set_ylabel('Efficiency (higher is better)')
    ax.set_xlabel('')
    plt.tight_layout()
    plt.rcParams.update({
    'font.size': 14,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 12,
    'lines.linewidth': 2,
    'savefig.dpi': 300
})
    plt.show()


def plot_tradeoff(summary_df):
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(8, 6))

    sns.scatterplot(
    data=summary_df,
    x='Clustering time (mean)',
    y='silhouette (mean)',
    hue='Algorithm',
    s=200,
    edgecolor='black',
    ax=ax)

    ax.set_title('Trade-off: Clustering Time vs. Quality', fontsize=16, weight='bold')
    ax.set_xlabel('Clustering Time (s) ')
    ax.set_ylabel('Silhouette ')
    plt.legend(title='Algorithm')
    plt.tight_layout()
    plt.rcParams.update({
    'font.size': 16,
    'axes.labelsize': 16,
    'axes.titlesize': 16,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'legend.fontsize': 16,
    'lines.linewidth': 2,
    'savefig.dpi': 300
})
    plt.show()


if __name__ == "__main__":
    filepath = "/home/sweksha/ros_ws/src/swarm_aggregation/scripts/cluster_metrics.txt"
    summary_df, trial_df = parse_swarm_summary(filepath)

    trial_df = calculate_efficiency(trial_df)
    summary_df['Clustering Efficiency'] = (summary_df['silhouette (mean)'] / summary_df['Clustering time (mean)'])

    print("\n Parsed algorithms:", summary_df['Algorithm'].tolist())
    print("\nSummary DataFrame:\n", summary_df[['Algorithm', 'silhouette (mean)', 'Clustering time (mean)']])

    plot_efficiency_summary(summary_df)
    plot_tradeoff(summary_df)
```
