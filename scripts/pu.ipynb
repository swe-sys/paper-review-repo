```python
def split_list(input_list):
    sublists = []
    sublist = []
    prev_num = None

    for num in input_list:
        if prev_num is None or num - prev_num <= 3:
            sublist.append(num)
        else:
            sublists.append(sublist)
            sublist = [num]
        prev_num = num

    sublists.append(sublist)
    return sublists


# Example usage
input_list1 = [19, 20, 21, 22, 23, 24, 246, 247, 248, 249, 250]
output1 = split_list(input_list1)
print(output1)

input_list2 = [19, 20, 21, 22, 23, 24, 50, 51, 52, 53, 54, 55, 246, 247, 248, 249, 250]
output2 = split_list(input_list2)
print(output2)
```
